##Task 3
###problem statement
**Explain the use of grid-auto-row and grid-auto-column using code examples.**

###grid-auto-rows

 * **(initial value - auto)**
grid-auto-rows property specifies the height of rows that are automatically created when there is no explicit
row definition.


###grid-auto-column

 * **(initial value - auto)**
 Similarly, the grid-auto-columns property specifies the width of columns that are automatically created when
there is no explicit column definition.


```
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Grid Auto Rows & Columns</title>
  <style>
    .container {
      display: grid;
      grid-template-columns: 100px 100px;  
      grid-template-rows: 100px 100px;     
      grid-auto-columns: 120px;           
      grid-auto-rows: 80px;                
      gap: 10px;
      border: 2px solid black;
      padding: 10px;
    }

    .item {
      background-color: lightseagreen;
      color: white;
      font-size: 18px;
      font-family: sans-serif;
      display: flex;
      justify-content: center;
      align-items: center;
      border-radius: 5px;
    }

    /* Kuch items ko extra row/column mein bhej rahe hain */
    .item5 {
      grid-column: 3; /* Ye column define nahi thi, auto banegi */
      grid-row: 1;
    }

    .item6 {
      grid-row: 3; /* Ye row define nahi thi, auto banegi */
      grid-column: 1;
    }

  </style>
</head>
<body>

  <div class="container">
    <div class="item">1</div>
    <div class="item">2</div>
    <div class="item">3</div>
    <div class="item">4</div>
    <div class="item item5">5</div>
    <div class="item item6">6</div>
    <div class="item item7">7</div>
  </div>

</body>
</html>


```

##Task 5

 ###Problem Statement

 **Explain the difference between justify-items and justify-self using code examples.**


**justify-items**

* This property aligns the items horizontally within the grid cell. It controls how the content within
a cell is aligned horizontally.

**Example**

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <style>
      .container {
        display: grid;
        grid-template-rows: repeat(2, 60px);
        grid-template-columns: repeat(6, 60px);
        gap: 10px;
        padding: 20px;
        border: 2px solid black;
        width: max-content;
        justify-items:  center;
      
      }
      .box {
        background-color: grey;
        border-radius: 2px;
        display: flex;
        justify-content: center;
        align-items: center;
        
      }
     .dark{
        background-color: rgb(70, 70, 70);
        color: white;
      }
      .box1{
        grid-row: 2;
    
        grid-column: 1;
      }
      .box2{
        grid-row: 2;
        grid-column: 2;
      }
    </style>
  </head>
  <body>
    <div class="container">
      <div class="box box1 dark">1</div>
      <div class="box box2">2</div>
      <div class="box box3 dark">3</div>
      <div class="box box4">4</div>
      <div class="box box5 dark">5</div>
      <div class="box box6">6</div>
      <div class="box box7 dark">7</div>
      <div class="box box8">8</div>
    </div>
  </body>
</html>

```

**justify-self**

* The CSS justify-self property sets the way a box is justified inside its alignment container along the appropriate axis.


**Example**

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <style>
      .container {
        display: grid;
        grid-template-rows: repeat(2, 60px);
        grid-template-columns: repeat(6, 60px);
        gap: 10px;
        padding: 20px;
        border: 2px solid black;
        width: max-content;
      
      }
      .box {
        background-color: grey;
        border-radius: 2px;
        display: flex;
        justify-content: center;
        align-items: center;
        justify-self: end;
        
        
      }
     .dark{
        background-color: rgb(70, 70, 70);
        color: white;
      }
      .box1{
        grid-row: 2;
    
        grid-column: 1;
      }
      .box2{
        grid-row: 2;
        grid-column: 2;
      }
    </style>
  </head>
  <body>
    <div class="container">
      <div class="box box1 dark">1</div>
      <div class="box box2">2</div>
      <div class="box box3 dark">3</div>
      <div class="box box4">4</div>
      <div class="box box5 dark">5</div>
      <div class="box box6">6</div>
      <div class="box box7 dark">7</div>
      <div class="box box8">8</div>
    </div>
  </body>
</html>

```

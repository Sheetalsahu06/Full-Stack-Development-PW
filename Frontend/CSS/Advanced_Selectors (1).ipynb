**Q2. Create a CSS file and link it to the provided HTML file below, use any of the combinator class selectors to style the navigation menu as follows-**

* Use the Descendant combinator selector to remove list style type bullet points and text underlining of the Menu list items
* Use the child combinator to display the menu list item inline
* Use the Adjacent sibling combinator selector to apply some margin between the menu list items.

```
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Styled Navigation Menu</title>
    <style>
        <!-- using descendant selector  remove list style type bullet points and text underlining   -->
        ul>li,
        li>a {
            list-style: none;
            text-decoration: none;
        }
        <!-- target child elements of ul using child combinator -->
        ul>li {
            display: inline;

        }
        <!--  target li using Adjacent sibling combinator selector -->
        ul li + li {
            margin:  0 20px;
        }

    </style>
</head>

<body>
    <header>
        <nav>
            <ul>
                <li><a href="#">Home</a></li>
                <li><a href="#">About</a></li>
                <li><a href="#">Services</a></li>
                <li><a href="#">Portfolio</a></li>
                <li><a href="#">Contact</a></li>
            </ul>
        </nav>
    </header>
    <main>
        <!-- your content here -->
    </main>
</body>

</html>
```

**Q4. Using any of the Pseudo element selectors, add any random words after and before the given HTML tag.**

* An example should be one HTML button tag.
*  When hovering the color should be changed and the shadow of the button should be added.
* Proper padding space should be given.
* The cursor style should also be changed to a pointer on hover.


```
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Button</title>
    <style>
        body {
            padding: 0;
            margin: 0;
            font-family: Arial, sans-serif;
        }

        .container {
            display: flex;
            column-gap: 100px;
            box-sizing: border-box;
            margin-top: 30px;
            justify-content: center;

        }

        .btn-container {
            margin: 25px;
            text-align: center;

        }

        .btn-container>label {
            display: block;
            margin-bottom: 8px;

        }

        .btn {
            padding: 20px 40px;
            background-color: rgb(58, 58, 255);
            color: white;
            border: none;
            border-radius: 10px;
            font-size: 16px;
            cursor: pointer;
            font-size: 20px;
            position: relative; /*equired for ::before and ::after pseudo-elements*/
        }

        .btn:hover {
            background-color: red;
            cursor: pointer;
            box-shadow: 10px 10px 10px rgba(36, 36, 36, 0.5);


        }

        .btn::before {
            content: "<<";
            position: absolute; /*required for positioning */
            left: -40px;
            font-size: 25px;
            color: black;
        }
        .btn::after {
            content: ">>";
            font-size: 25px;
            position: absolute; /*required for positioning */
            right: -40px;
            color: black;

            
        }

      
    </style>
</head>

<body>
    <div class="container">
        <div class="btn-container">
            <label>Without hover</label>
            <button class="btn">PW Skills</button>
        </div>
        <div class="btn-container">
            <label>On hover</label>
            <button class="btn">PW Skills</button>
        </div>
    </div>
</body>

</html>

```

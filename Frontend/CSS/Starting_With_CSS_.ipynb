**1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required.**


```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CSS Example</title>
    
    <!-- Internal CSS -->
    <style>
        /* Internal CSS */
        .box {
            width: 250px;
            height: 250px;
            background-color: lightblue;
            margin: 10px;
            padding: 20px;
            text-align: center;
        }
    </style>
    
    <!-- External CSS Link -->
    <link rel="stylesheet" href="styles.css">
</head>
<body>
    <!-- Inline CSS -->
    <div class="box" style="color: rgb(255, 0, 0);">
        <p>Box 1</p>
    </div>
    
    <div class="box">
        <p>Box 2</p>
    </div>
    <div class="box">
        <p>Box 3</p>
    </div>
</body>
</html>
```


styles.css

```
/* External CSS */
.box {
    background-color: rgb(143, 200, 143);
    color: rgb(0, 0, 0);
}

```

**2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.**

```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BEM Naming Convention Example</title>
    
    <!-- Internal CSS using BEM Naming Convention -->
    <style>
        /* Block: paragraph */
        .paragraph--intro {
            font-size: 18px;
            color: darkblue;
            font-weight: bold;
        }
        
        .paragraph--highlight {
            font-size: 16px;
            background-color: yellow;
            padding: 5px;
        }
        
        .paragraph--note {
            font-size: 14px;
            font-style: italic;
            color: gray;
        }
    </style>
</head>
<body>
    <h2>BEM Naming Convention Example</h2>
    
    <!-- Applying BEM naming convention to paragraphs -->
    <p class="paragraph--intro">This is an introduction paragraph with emphasized styling.</p>
    <p class="paragraph--highlight">This paragraph is highlighted to draw attention.</p>
    <p class="paragraph--note">This is a note paragraph styled with a different font and color.</p>
</body>
</html>

```

**3.Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.**

```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Styled HTML Form</title>
    
    <!-- Internal CSS with Custom Color Palette and Opacity -->
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #f4f4f4;
            padding: 20px;
        }
        
        .form-container {
            width: 300px;
            background-color: #ffffff;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
        }
        
        .form-section {
            margin-bottom: 15px;
        }
        
        .form-section--opaque {
            opacity: 0.7; /* Applying opacity to this section */
        }
        
        label {
            font-weight: bold;
            display: block;
            margin-bottom: 5px;
        }
        
        input, select, textarea {
            width: 100%;
            padding: 8px;
            margin-bottom: 10px;
            border: 1px solid #ccc;
            border-radius: 5px;
        }
        
        input[type="text"] {
            background-color: #e0f7fa; /* Light cyan */
        }
        
        input[type="email"] {
            background-color: #ffecb3; /* Light yellow */
        }
        
        textarea {
            background-color: #fce4ec; /* Light pink */
        }
        
        button {
            background-color: #007BFF;
            color: white;
            padding: 10px;
            border: none;
            border-radius: 5px;
    
        }
        
    </style>
</head>
<body>
    <h2>Styled Form with Custom Colors</h2>
    
    <div class="form-container">
        <form>
            <div class="form-section">
                <label for="name">Name:</label>
                <input type="text" name="name" required>
            </div>
            
            <div class="form-section--opaque">
                <label for="email">Email:</label>
                <input type="email" name="email" required>
            </div>
            
            <div class="form-section">
                <label for="message">Message:</label>
                <textarea  name="message" rows="4" required></textarea>
            </div>
            
            <button type="submit">Submit</button>
        </form>
    </div>
</body>
</html>
```
